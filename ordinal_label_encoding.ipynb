```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('customer.csv')
```

```python
df.sample(5)
```

```python
df = df.iloc[:,2:]
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:2],
                                                 df.iloc[:,-1],test_size=0.2)
```

```python
X_train
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
#ordinal encoding will be performed because these columns have ordinal data
oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
```

```python
oe.fit(X_train)
```

```python
X_train = oe.transform(X_train)
X_test = oe.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
oe.categories_
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
#Label encoding is performed as it is output column and have categorical data
#no parameters so it'll decide random values
le = LabelEncoder()
```

```python
le.fit(y_train)
```

```python
le.classes_
```

```python
y_train = le.transform(y_train)
y_test = le.transform(y_test)
```
